```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat


b9_data = loadmat('blocc_9node.mat', squeeze_me=True)
b3_data = loadmat('blocc_3node.mat', squeeze_me=True)
s9_data = loadmat('sflcb_9node.mat', squeeze_me=True)
s3_data = loadmat('sflcb_3node.mat', squeeze_me=True)


b9_his_time = b9_data['his_time']
b9_obj = b9_data['obj_vec']
b3_his_time = b3_data['his_time']
b3_obj = b3_data['obj_vec']


s9_his_time = s9_data['his_time']
s9_obj = s9_data['his_ob_z']
s3_his_time = s3_data['his_time']
s3_obj = s3_data['his_ob_z']

plt.figure(figsize=(6, 4))




plt.plot(b9_his_time, b9_obj, label='BLOCC', linewidth=2)
plt.plot(s9_his_time, s9_obj, label='SFLCB', linewidth=2)

plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('UL Utility', fontsize=14)



plt.xlim(0, 400)
plt.ylim(-100, 25)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()

plt.savefig('9node.png', dpi=300)
plt.show()

plt.figure(figsize=(6, 4))


plt.plot(b3_his_time, b3_obj, label='BLOCC', linewidth=2)
plt.plot(s3_his_time, s3_obj, label='SFLCB', linewidth=2)

plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('UL Utility', fontsize=14)


plt.xlim(0, 5)
plt.ylim(0, 2)
plt.legend(fontsize=12)
plt.grid(True)
plt.tight_layout()

plt.savefig('3node.png', dpi=300)
plt.show()


```
